```python
import matplotlib.pyplot as plt
flood_years = [1960, 1970, 1980, 2014, 2017, 2019, 2022]
drought_years = [1990, 2003, 2004, 2016, 2017, 2018, 2024]
flood_labels = ['Encroachment', 'Encroachment', 'Encroachment', 'Heavy Rainfall', 'Intense Rainfall', 'Heavy Rains', 'Flooding']
drought_labels = ['Water Shortages', 'Severe Water Crisis', 'Severe Water Crisis', 'Major Drought', 'Below-average Rainfall', 'Below-average Rainfall', 'Extreme Conditions']
plt.figure(figsize=(14, 7))
plt.style.use('seaborn-darkgrid')
for i, year in enumerate(flood_years):
    plt.scatter(year, 1, color='blue', s=100, edgecolors='w', linewidth=2, label='Floods' if i == 0 else "")
    plt.text(year, 1.1, flood_labels[i], rotation=45, ha='right', fontsize=10, color='blue')

# Drought events
for i, year in enumerate(drought_years):
    plt.scatter(year, 0, color='red', s=100, edgecolors='w', linewidth=2, label='Droughts' if i == 0 else "")
    plt.text(year, -0.1, drought_labels[i], rotation=45, ha='right', fontsize=10, color='red')

# Adding titles and labels
plt.title('Flood and Drought Events Over the Years', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.yticks([0, 1], ['Droughts', 'Floods'], fontsize=12)
plt.ylim(-1, 2)
plt.xlim(min(min(flood_years), min(drought_years)) - 5, max(max(flood_years), max(drought_years)) + 5)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def get_gpt_response(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.7,
    )
    return response.choices[0].text.strip()
```

```python
def preprocess_input(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)
```
